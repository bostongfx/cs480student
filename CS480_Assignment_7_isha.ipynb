```python
# TODO: YOUR ANSWER
```

MPR is an acronym for Multiplanar resconstruction or reformation (MPR) which is the process of using the data from axial CT images to create nonaxial two-dimensional images. It allows images to be created from the original axial plane in either the coronal, sagittal, or oblique plane

```python
# Question 2): What is the difference between MIP, MinIP, AvgIP?
#
```

```python
# TODO: YOUR ANSWER
```

They are all volume-rendering techniques in which suitable editing methods are used to define the volume of interest.

-> MIP (Maximum Intensity Projection) method tends to display bone and contrast material-filled structures preferentially, and other lower-attenuation structures are not well visualized MIP images are not threshold dependent and preserve attenuation information.

-> MinIP (Minimum Intensity Projection) enables detection of low-density structures in a given volume. it consists of projecting the voxel with the lowest attenuation value on every view throughout the volume onto a 2D image.

-> Avg (Average Intensity Projection) operates in a similar way to the other modes of intensity projection, but instead of showing voxels with the highest/lowest attenuation, shows the average attenuation.
